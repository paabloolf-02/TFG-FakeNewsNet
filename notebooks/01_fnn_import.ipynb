#Cargar los csv

```python
import pandas as pd

# Rutas de los CSV originales
fake_path = f"{PROJECT_DIR}/data/Fake.csv"
true_path = f"{PROJECT_DIR}/data/True.csv"

# Me aseguro que dentro de data estan los csv que quiero
!ls data

# Cargo los datasets
df_fake = pd.read_csv(fake_path)
df_true = pd.read_csv(true_path)

print("df_fake:", df_fake.shape)
print("df_true:", df_true.shape)

df_fake.head()
```

```python
# Etiquetas: fake = 1, true = 0
df_fake["label"] = 1
df_true["label"] = 0

# Uno los dos dataframes
df = pd.concat([df_fake, df_true])

# Mezclo las filas para evitar orden sesgado
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

print("Dataset final unificado:", df.shape)

df.head()
```

```python
# Guardo el dataset unificado (sin limpiar)
output_path = f"{PROJECT_DIR}/data/noticias.csv"
df.to_csv(output_path, index=False)

print("Guardado en:", output_path)
```
