```python
from getpass import getpass

token = getpass("Introduce tu token personal de GitHub: ")

# Construyo la URL temporal con usuario + token
remote_with_token = f"https://paabloolf-02:{token}@github.com/paabloolf-02/TFG-FakeNewsNet.git"

print("Configurando remoto con token temporal...")
# Cambio la URL del remoto 'origin' para este push
!git remote set-url origin "$remote_with_token"

print("Subiendo cambios a GitHub...")
# Hago el push a la rama main
!git push origin main

print("Restaurando URL original sin token...")
# Dejo el remoto como debe quedar normalmente (sin credenciales)
!git remote set-url origin "https://github.com/paabloolf-02/TFG-FakeNewsNet.git"

print("Push completado correctamente.")

```
